```python
import pandas as pd
path=r"C:\Users\user\Downloads\airbnb.csv\airbnb.csv"
airbnb=pd.read_csv(path)
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x=airbnb[['price','number_of_reviews']]
x_scaled=ss.fit_transform(x)

x_scaled[:,0],x_scaled[:,1]
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_scaled[:,0],x_scaled[:,1]) #第0行、第1行
plt.xlabel('price')
plt.ylabel('number of reviews')
```

```python
from sklearn.cluster import KMeans 
from sklearn.metrics import silhouette_score 
silhouette=[]
for i in [2,3,4,5]:
    kmeans=KMeans(n_clusters=i)
    y_pred=kmeans.fit_predict(x_scaled)
    silhouette.append(silhouette_score(x_scaled,y_pred))
```

```python
a=[2,3,4,5]
plt.plot(a,silhouette)
plt.xlabel('k cluster')
plt.ylabel('silhouette_score')
plt.grid()
```

```python
kmeans=KMeans(n_clusters=4)
y_pred=kmeans.fit_predict(x_scaled)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_scaled[:,0],x_scaled[:,1],c=y_pred) #第0行、第1行
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.grid()
```
